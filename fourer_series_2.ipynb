## フーリエ級数展開
以下の周期関数を三角関数の無限級数の形式を考える．
$$
\begin{align*}
&f(t) = t \quad (-1\leq t < 1)
\\
&f(t+2) = f(t)
\end{align*}
$$

```python
fig, ax = plt.subplots(figsize=(13, 3))
plot_periodic_ex2(ax)
plt.tight_layout()
plt.show()
```

#### Step1: フーリエ係数を計算する
$$
\begin{align*}
C_n &= \frac{1}{2L} \int^L_{-L} f(t) \exp \left[-i \frac{n\pi}{L}t\right] dt
\\\\
&=
\begin{cases}
\displaystyle
0 \quad \textrm{if}\quad n= 1
\\\\
\displaystyle
\frac{i}{n\pi} (-1)^n \quad \textrm{if}\quad n\neq 1
\end{cases}
\end{align*}
$$

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 4))
plot_periodic_ex2(ax[0])
plot_Cn_ex2(ax[1])
plt.tight_layout()
plt.show()
```

係数：重ね合わせる波動の振幅と考えて良い．

#### Step2: フーリエ級数展開を計算する
$$
\begin{align*}
f(t) &= \sum^\infty_{n=-\infty} C_n \exp \left[in\pi t \right]
\qquad \left( = \frac{1}{2} + \sum^{\infty}_{k=1} \frac{2}{(2k+1)\pi} \sin \left[(2k+1)\pi t \right]\right)
\end{align*}
$$
実際にはコンピューターは$\infty$を扱えないので，$n\leq |N|$で総和を打ち切ってフーリエ級数展開を計算する：
$$
\begin{align*}
f(t) &= \sum^N_{n=-N} C_n \exp \left[in\pi t \right]
\end{align*}
$$

```python
N = 1
fig, ax = plt.subplots(1, 2, figsize=(16, 4))
plot_Cn_ex2(ax[0], N, N)
plot_periodic_reconst_ex2(ax[1], N)
plt.tight_layout()
plt.show()
```
